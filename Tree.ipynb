```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
import pyroc
from mpl_toolkits.mplot3d import Axes3D
from sklearn import tree
```

```python
%matplotlib notebook
```

```python
df=pd.read_csv("./winequality-red.csv")
```

```python
df.loc[df.quality > 5, 'category'] = 1
df.loc[df.quality <= 5, 'category'] = 0
```

```python
clf=tree.DecisionTreeClassifier()
```

```python
df_train=df.iloc[:500,]
df_test=df.iloc[500:,]
```

```python
X=pd.DataFrame()
X['alcohol']=df_train['alcohol']
X['sulphates']=df_train['sulphates']
X['chlorides']=df_train['chlorides']
y=df_train['category']
```

```python
X2=pd.DataFrame()
X2['alcohol']=df_test['alcohol']
X2['sulphates']=df_test['sulphates']
X2['chlorides']=df_test['chlorides']
y2=df_test['category']
```

```python
clf=clf.fit(X,y)
```

```python
clf.score(X,y)
```

```python
X2
```

```python
clf.score(X2,y2)
```

```python
X2['predicted']=clf.predict(X2)
X2
```

```python
fig = plt.figure()#.gca(projection='3d')
ax2 = fig.add_subplot(111, projection='3d')
col=['g','r']
for i in range(0,1099):
    ax2.scatter(X2.iloc[i]['alcohol'], X2.iloc[i]['sulphates'], X2.iloc[i]['chlorides'], zdir='z', s=20, \
                c=col[int(X2.iloc[i]['predicted'])], depthshade=True)
```

```python
from sklearn.metrics import precision_recall_fscore_support
y_true = df_test['category']
y_pred = X2['predicted']
precision_recall_fscore_support(y_true, y_pred, average='binary')
```
